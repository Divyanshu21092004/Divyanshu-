# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python
df.shape
```

```python

# Check for missing values and data types
# Check data types and non-null counts
df.info()

# Check for missing values
print("\nMissing values in Bank_marketing_inspection data set in  each column:\n")
print(df.isnull().sum())
```

```python
df.dropna()
```

```python

# Summary statistics
df.describe()
```

```python
df.describe(include='object')
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
sns.countplot(data=df, x='y', palette='Set2')
plt.title('Distribution of Term Deposit Subscription')
plt.xlabel('Subscribed to Term Deposit')
plt.ylabel('Number of Clients')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the counts
labels = df['y'].value_counts().index
sizes = df['y'].value_counts().values
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)  # "explode" the first slice slightly

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=colors, explode=explode, shadow=True)
plt.title('3D-like Pie Chart - Term Deposit Subscription')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
percent = df['y'].value_counts(normalize=True) * 100
print(percent)
```

```python

# Correlation matrix for numerical features
corr_matrix = df.corr(numeric_only=True)

# Display the correlation matrix
print(corr_matrix)
```

```python
df['y_numeric'] = df['y'].map({'no': 0, 'yes': 1})
```

```python
correlations = df.corr(numeric_only=True)['y_numeric'].sort_values(ascending=False)
print(correlations)
```

```python
job_impact = df.groupby('job')['y'].value_counts(normalize=True).unstack()
print(job_impact)

# Visualization


sns.barplot(data=df, x='job', y='y_numeric', estimator='mean')
plt.xticks(rotation=45)
plt.title('Subscription Rate by Job')
plt.ylabel('Proportion of "Yes" Responses')
plt.show()
```
